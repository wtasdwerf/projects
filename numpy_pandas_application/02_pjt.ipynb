```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def file_open_by_numpy():
    # np.loadtxt(구분자 = ',', 데이터 타입: string)
    np_arr = np.loadtxt('data/NFLX.CSV', delimiter=",", encoding='cp949', dtype=str)
    return np_arr
```

```python
arr = file_open_by_numpy()
print(arr)
```

```python
df = pd.DataFrame(arr)
df
```

```python
columns=arr[0]
arr = np.delete(arr, 0, 0)
df = pd.DataFrame(arr, columns=columns)
df
```

```python
df.loc[:, 'Date':'Close']
```

```python
df['datetime'] = pd.to_datetime(df['Date'])
```

```python
# CSV 파일 경로
csv_path = "data/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# 2021년 이후 데이터 필터링
df_after_2021 = df[df["Date"] >= "2021-01-01"]

df_after_2021
```

```python
# 날짜 데이터 변환
df_after_2021["Date"] = pd.to_datetime(df_after_2021["Date"])

# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df_after_2021['Date'], df_after_2021['Close'])


# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Close')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)


# 그래프 표시
plt.show()
```

```python
max_price = df_after_2021['Close'][954] # idxmax()
min_price = df_after_2021['Close'][1001] # idxmin()


print("최고 종가:", max_price)
print("최저 종가:", min_price)
```

```python
df_after_2021['new_frame'] = df_after_2021['Date'].dt.strftime('%Y-%m')

df_after_2021.head()
```

```python
df_new = df_after_2021.groupby(['new_frame']).mean()
df_new
```

```python
# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df_new['Date'], df_new['Close'])

# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Close')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)


# 그래프 표시
plt.show()
```

```python
df_new2 = df_after_2021[df_after_2021["Date"] >= "2022-01-01"]

# 그래프 그리기
plt.plot(df_new2['Date'], df_new2['Close'], label='Close')
plt.plot(df_new2['Date'], df_new2['Low'], label='Low')
plt.plot(df_new2['Date'], df_new2['High'], label='High')

# 그래프 제목 설정
plt.title('High, Low Prices over Time')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Price')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```
